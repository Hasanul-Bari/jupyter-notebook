```python
import matplotlib.pyplot as mpl
%matplotlib inline
```

```python
x=[2,4,6,8,10]
y=[50,65,20,25,80]

mpl.plot(x,y, color='cyan', linestyle='dotted', linewidth=5)

mpl.xlabel('Item')
mpl.ylabel('Price')
mpl.title('Price-chart')
```

```python
x=[2,4,6,8,10]
y=[50,65,20,25,80]

mpl.scatter(x,y)
```

```python
x=[2,4,6,8,10]
y=[50,65,20,25,80]

mpl.plot(x,y,'g+')
```

```python
days=[1,2,3,4,5,6]
mx=[40,45,64,50,55,60]
mn=[20,25,30,27,28,40]
av=[30,35,47,38,42,50]


mpl.plot(days,mx, label='Max')
mpl.plot(days,mn, label='Min')
mpl.plot(days,av, label='Avg')

mpl.legend(loc='best', shadow=True)
mpl.grid()
```

```python
import numpy as np

product=np.array([1,2,3,4])
price=[15,25,40,30]
profit=[5,7,10,5]

mpl.bar(product-.2,price, label='price', color='red', width=.4)
mpl.bar(product+.3,profit, label='profit', color='blue', width=.4)
mpl.legend()
```

```python
price=[15,20,25,30,31,31,32,32,32,33,34,35,40,42,42,45,45,48,45,44,60]
mpl.hist(price)
```

```python
val=[300,500,1600,450,360]
lab=['pen','notebook','pencil','phone','calculator']

mpl.pie(val, labels=lab, radius=2, autopct='%.2f%%', startangle=45, explode=[0,.2,0,.1,0])
```
